```python
import pandas as pd
movies = pd.read_csv('movies.csv', sep=',', quotechar='"', on_bad_lines='skip');
display(movies.head(5))
```

```python
movies = movies.rename(columns={'index' : 'user_id'})
movies = movies.rename(columns={'id' : 'movies_id'})
movies = movies.rename(columns={'vote_average' : 'rating'})
display(movies.head(7))
```

```python
#I delete my not nasasary rows...
movies = movies.drop('director', axis=1)

# Display the DataFrame after dropping columns
display(movies.head(5))
```

```python
display(movies.head(7))
```

```python
ratings = pd.read_csv('Movie-Dataset-Latest.csv')
display(ratings.head(5))
```

```python
ratings = ratings.rename(columns={'Unnamed: 0' : 'user_id'})
ratings = ratings.rename(columns={'id' : 'movie_id'})
ratings = ratings.rename(columns={'vote_average' : 'ratings'})

display(ratings.head(5))
```

```python
#I delete not nassasary rows...
ratings = ratings.drop('vote_count', axis=1)

# Display the DataFrame after dropping columns
display(ratings.head(5))
```

```python
display(ratings.head(7))
```

```python
user_item_matrix = ratings.pivot(index='user_id', columns='movie_id', values='ratings').fillna(0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#calculate user similarity
user_similarity = cosine_similarity(user_item_matrix)
user_similarity_df = pd.DataFrame(user_similarity, index=user_item_matrix.index, columns=user_item_matrix.index)
```

```python
count_vectorizer  = CountVectorizer()
genre_matrix = count_vectorizer.fit_transform(movies['genres'].fillna(''))
```

```python
genre_similarity = cosine_similarity(genre_matrix)
genre_similarity_df = pd.DataFrame(genre_similarity, index=movies['movies_id'], columns=movies['movies_id'])
```

```python
# Recommendation function
def recommend_movies(user_id, user_item_matrix, user_similarity_df, genre_similarity_df, num_recommendations=3):
    # Get the user's ratings
    user_ratings = user_item_matrix.loc[user_id]

    # Get similar users
    similar_users = user_similarity_df[user_id].sort_values(ascending=False)
    similar_users = similar_users[similar_users.index != user_id].head(num_recommendations)

    # Get movie recommendations based on similar users' ratings
    recommended_movies = {}

    # Get valid movie IDs from the movies DataFrame
    valid_movie_ids = movies['movies_id'].unique()

    for similar_user in similar_users.index:
        similar_user_ratings = user_item_matrix.loc[similar_user]
        for movies_id, rating in similar_user_ratings.items():
            if rating > 0 and movies_id not in user_ratings.index[user_ratings > 0] and movies_id in valid_movie_ids:
                if movies_id not in recommended_movies:
                    recommended_movies[movies_id] = 0
                recommended_movies[movies_id] += rating * similar_users[similar_user]

         # Sort recommendations by score
    recommended_movies = sorted(recommended_movies.items(), key=lambda x: x[1], reverse=True)

    # Get movie titles for recommended movie IDs
    recommended_movie_titles = []
    for movies_id, _ in recommended_movies[:num_recommendations]:
        # No need to check if movie_id exists in movies anymore as we filtered based on valid_movie_ids
        recommended_movie_titles.append(movies[movies['movies_id'] == movies_id]['title'].values[0])


    return recommended_movie_titles
# Example usage
user_id = 1
recommended_movies = recommend_movies(user_id, user_item_matrix, user_similarity_df, genre_similarity_df)
print(f"Recommended movies for user {user_id}: {recommended_movies}")
```
